# **Introducción a PIL/Pillow**
### **Historia y Contexto:**

**PIL (Python Imaging Library):** Desarrollada originalmente por Fredrik Lundh, PIL fue una de las primeras bibliotecas de procesamiento de imágenes para Python. Permitió a los desarrolladores realizar operaciones básicas y avanzadas en imágenes de manera eficiente.

**Pillow:** En 2011, Pillow surgió como una bifurcación amigable y mantenida de PIL para garantizar la compatibilidad con las versiones más recientes de Python y agregar nuevas características. Actualmente, Pillow es la biblioteca estándar de facto para el procesamiento de imágenes en Python.

# **PIL**
* **Carga y Guardado de Imágenes:**Soporta múltiples formatos de imagen como JPEG, PNG, GIF, BMP, y más.
* **Operaciones Básicas:** Redimensionamiento, recorte, rotación, y conversión de imágenes a diferentes modos de color.
* **Filtros y Efectos:** Aplicación de filtros como desenfoque, contorno, suavizado, y más.

### **Aplicaciones Comunes:**

Creación de miniaturas, conversión de formatos de imagen, adición de efectos simples a imágenes.

# **Ejemplo**

```python
from PIL import Image, ImageFilter

# Cargar una imagen
img = Image.open("/content/Procesamiento-Digital-de-Imagenes.jpg")

# Redimensionar la imagen
img_resized = img.resize((300, 300))

# Aplicar un filtro de desenfoque
img_blurred = img_resized.filter(ImageFilter.BLUR)

# Guardar la imagen resultante
img_blurred.save("imagen_blurred.jpg")
```

# **OpenCV**
**Descripción:**

OpenCV (Open Source Computer Vision Library) es una biblioteca de visión por computadora y aprendizaje automático, diseñada para proporcionar una infraestructura común para aplicaciones de visión.

### **Características Principales:**

**Operaciones Básicas:** Redimensionamiento, recorte, rotación, conversión de color, y transformaciones geométricas.

**Procesamiento de Imágenes:**Detección de bordes, umbralización, filtrado, y transformaciones morfológicas.

**Detección de Objetos:** Detección de caras, ojos, y otros objetos utilizando clasificadores pre-entrenados.

**Funciones Avanzadas:** Segmentación, reconocimiento de patrones, y seguimiento de objetos.

**Aplicaciones Comunes:**
Procesamiento en tiempo real de video, detección de objetos, reconocimiento de patrones, aplicaciones de realidad aumentada.

# **Ejemplo**

```python
import cv2

# Cargar una imagen
img = cv2.imread("/content/Procesamiento-Digital-de-Imagenes.jpg")

# Convertir a escala de grises
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Aplicar detección de bordes
edges = cv2.Canny(gray, 100, 200)

# Guardar la imagen resultante
cv2.imwrite("imagen_edges.jpg", edges)
```

```python
import cv2

# Cargar la imagen
img = cv2.imread("/content/Procesamiento-Digital-de-Imagenes.jpg")

# Convertir a escala de grises
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Aplicar un filtro de detección de bordes
edges = cv2.Canny(img_gray, 100, 200)

# Dibujar un círculo en el centro de la imagen
(h, w) = img.shape[:2]
center_x, center_y = w // 2, h // 2
cv2.circle(edges, (center_x, center_y), 50, (255, 255, 255), 3)

# Dibujar un texto en la parte inferior de la imagen
cv2.putText(edges, "Detección de Bordes", (10, h - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

# Guardar la imagen resultante
cv2.imwrite("imagen_procesada.jpg", edges)

# Mostrar la imagen resultante
cv2.imshow("Imagen Procesada", edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# **Scikit-Image**

### **Descripción:**

Scikit-Image es una biblioteca de procesamiento de imágenes en Python que forma parte del ecosistema SciPy. Está diseñada para ser simple y fácil de usar.

**Características Principales:**

**Filtros y Transformaciones:**
Desenfoque, afilado, transformaciones de Fourier, y más.

**Segmentación y Etiquetado:**
Técnicas avanzadas de segmentación, etiquetado de regiones, y detección de contornos.

**Medición y Análisis:**
Cálculo de propiedades geométricas y estadísticas de regiones de la imagen.

**Operaciones Geométricas:**
Redimensionamiento, rotación, traslación, y deformaciones.

**Aplicaciones Comunes:**
Análisis y medición de imágenes científicas, preprocesamiento de imágenes para aprendizaje automático, desarrollo de algoritmos de segmentación personalizados.

# **Ejemplo**

```python
from skimage import io, filters, color
import numpy as np

# Cargar una imagen
img = io.imread("/content/Procesamiento-Digital-de-Imagenes.jpg")

# Convertir a escala de grises
gray = color.rgb2gray(img)

# Aplicar un filtro de Sobel para detectar bordes
edges = filters.sobel(gray)

# Convertir la imagen a un formato compatible con PNG
edges_uint8 = (edges * 255).astype(np.uint8)

# Guardar la imagen resultante
io.imsave("/content/imagenGr.png", edges_uint8)

```

## **SimpleCV**

###**Descripción:**

SimpleCV es una biblioteca de visión por computadora que proporciona una interfaz sencilla para tareas complejas de procesamiento de imágenes y visión artificial.

### **Características Principales:**

**Operaciones Básicas:**
Carga, guardado, redimensionamiento, y conversión de imágenes.

**Detección y Seguimiento de Objetos:**
Detección de caras, ojos, y otros objetos. Seguimiento de objetos en secuencias de video.

**Análisis de Color:** Detección de colores específicos, histograma de colores, y operaciones en espacios de color.

**Integración con Hardware:**
Captura de video en tiempo real desde cámaras, integración con dispositivos de entrada/salida.

**Aplicaciones Comunes:**
Prototipado rápido de aplicaciones de visión artificial, proyectos educativos, desarrollo de aplicaciones interactivas basadas en visión.

# **Herramientas en Línea**
**Descripción:**

Existen diversas herramientas en línea que permiten realizar operaciones básicas de procesamiento de imágenes sin necesidad de instalar software adicional.
Ejemplos de Herramientas:

**Pixlr:**
Editor de imágenes en línea que ofrece funciones de recorte, redimensionado, ajuste de color, y aplicación de filtros.

**Fotor:**
Herramienta en línea para edición de fotos, creación de collages, y diseño gráfico.

**Canva:**
Plataforma de diseño gráfico que incluye funciones de edición de imágenes, creación de gráficos y presentaciones.

**Aplicaciones Comunes:**

Edición rápida de imágenes, creación de contenido visual para redes sociales y marketing, diseño gráfico básico.

# **Comparación y Selección de Herramientas**

##**Facilidad de Uso:**

* Herramientas como Pillow y SimpleCV son más accesibles para principiantes debido a su simplicidad y documentación clara.

* OpenCV y Scikit-Image, aunque más complejas, ofrecen mayor flexibilidad y potencia para usuarios avanzados.

## **Características y Funcionalidades:**

* OpenCV es ideal para aplicaciones en tiempo real y procesamiento de video.

* Scikit-Image es excelente para análisis y medición científica.

* SimpleCV es perfecto para prototipado rápido y proyectos educativos.

##**Integración y Extensibilidad:**

* OpenCV y Scikit-Image se integran bien con otros paquetes de Python y se pueden extender para aplicaciones personalizadas.

* Pillow es ideal para tareas básicas de procesamiento de imágenes y se puede combinar fácilmente con otras bibliotecas de Python.
